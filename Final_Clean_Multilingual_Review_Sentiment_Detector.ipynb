####Punctuation Removal Functions

```python
# Urdu-specific punctuation removal
URDU_PUNCTUATIONS = ['\200F', '\u200f', '۔', '٫', '٪', '؟', '،', ')', '(', '{', '}', '…', '۔۔۔', '\u002F', '\u003F']
def remove_punctuation_urdu(text):
    """
    Remove Urdu-specific punctuations from the text.
    """
    # UrduHack's punctuation removal
    text = urduhack_remove_punctuation(text)
    # Custom punctuation removal
    for punct in URDU_PUNCTUATIONS:
        text = text.replace(punct, " ")
    return text

# Punjabi-specific punctuation removal
PUNJABI_PUNCTUATIONS = ['،', '؟', '۔', ')', '(', '{', '}', '۔۔۔', '٪', '٫']
def remove_punctuation_punjabi(text):
    """
    Remove Punjabi-specific punctuations from the text.
    """
    for punct in PUNJABI_PUNCTUATIONS:
        text = text.replace(punct, " ")
    return text

# Siraiki-specific punctuation removal
# SIRAIKI_PUNCTUATIONS = ['،', '؟', '۔', ')', '(', '{', '}', '۔۔۔', '٪', '٫']
# def remove_punctuation_siraiki(text):
#     """
#     Remove Siraiki-specific punctuations from the text.
#     """
#     for punct in SIRAIKI_PUNCTUATIONS:
#         text = text.replace(punct, " ")
#     return text

# English-specific punctuation removal
def remove_punctuation_english(text):
    """
    Remove English-specific punctuations and numbers from the text.
    """
    return re.sub(f"[{string.punctuation}0-9]", " ", text)
```

####Language Specific Preprocessing

```python
# Language-specific preprocessing
def preprocess_urdu(text):
    """
    Preprocess Urdu text:
    - Normalize whitespace.
    - Remove punctuations, accents, URLs, emails, numbers, currency symbols, and English alphabets.
    """
    text = normalize_whitespace(text)
    text = urdu_remove_punctuation(text)
    text = remove_accents(text)
    text = replace_urls(text)
    text = replace_emails(text)
    text = replace_numbers(text)
    text = replace_currency_symbols(text)
    text = remove_english_alphabets(text)
    return text

def preprocess_punjabi(text):
    """
    Preprocess Punjabi text:
    - Normalize whitespace.
    - Remove URLs, emails, numbers, and punctuations.
    """
    text = normalize_whitespace(text)
    text = re.sub(r"http\S+|www.\S+", "", text)  # Remove URLs
    text = re.sub(r"\S+@\S+", "", text)  # Remove emails
    text = re.sub(r"\d+", "", text)  # Remove numbers
    text = remove_punctuation_punjabi(text)  # Remove Punjabi punctuations
    return text

# def preprocess_siraiki(text):
#     """
#     Preprocess Siraiki text:
#     - Normalize whitespace.
#     - Remove URLs, emails, numbers, and punctuations.
#     """
#     text = normalize_whitespace(text)
#     text = re.sub(r"http\S+|www.\S+", "", text)  # Remove URLs
#     text = re.sub(r"\S+@\S+", "", text)  # Remove emails
#     text = re.sub(r"\d+", "", text)  # Remove numbers
#     text = remove_punctuation_siraiki(text)  # Remove Siraiki punctuations
#     return text

def preprocess_english(text):
    """
    Preprocess English text:
    - Normalize whitespace.
    - Remove punctuations, numbers, URLs, and emails.
    """
    text = normalize_whitespace(text)
    text = re.sub(r"http\S+|www.\S+", "", text)  # Remove URLs
    text = re.sub(r"\S+@\S+", "", text)  # Remove emails
    text = re.sub(r"\d+", "", text)  # Remove numbers
    # text = re.sub(f"[{string.punctuation}]", " ", text)  # Remove general punctuations
    text = remove_punctuation_english(text)
    return text
```

####Text Cleaning Function

```python
def clean_text(text, language):
    """
    Clean text by applying language-specific preprocessing.
    Includes:
    - Removing hashtags, HTML tags, mentions, punctuations, and numbers.
    - Normalizing whitespace.
    """
    # Step 1: Remove hashtags, mentions, and HTML tags
    text = re.sub(r"#\w+", "", text)  # Remove hashtags
    text = re.sub(r"@\w+", "", text)  # Remove mentions
    text = re.sub(r"<.*?>", "", text)  # Remove HTML tags

    # Step 2: Language-specific preprocessing
    if language == "urdu":
        text = preprocess_urdu(text)
    elif language == "punjabi":
        text = preprocess_punjabi(text)
    # elif language == "siraiki":
    #     text = preprocess_siraiki(text)
    elif language == "english":
        text = preprocess_english(text)

    # Step 3: Normalize whitespace
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

####Lemmatizers for All Languages

```python
def lemmatize_english(text):
    """
    Lemmatize English text using NLTK's WordNetLemmatizer.
    """
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text)
    return " ".join([lemmatizer.lemmatize(word) for word in tokens])

def lemmatize_urdu(text):
    """
    Lemmatize Urdu text using UrduHack.
    """
    temp = urdu_lemmatizer.lemma_lookup(text)
    lemme_str = " ".join([t[0] for t in temp])
    return lemme_str

def lemmatize_punjabi(text):
    """
    Lemmatize Punjabi text. Placeholder function (no comprehensive Punjabi lemmatization library available).
    """
    # Example of a basic rule-based lemmatizer
    lemme_str = text.replace("اں", "").replace("اںو", "")
    return lemme_str

# def lemmatize_siraiki(text):
#     """
#     Lemmatize Siraiki text. Placeholder function (no comprehensive Siraiki lemmatization library available).
#     """
#     lemme_str = text.replace("واں", "").replace("اں", "")
#     return lemme_str
```

####Text Preprocessing Function

```python
# Define a preprocessing function for text data
def preprocess_text(text, language):
    """
    Preprocess the input text based on the specified language.
    - Removes punctuation, special characters, and numbers.
    - Converts to lowercase.
    - Tokenizes the text.
    - Removes stopwords based on language.
    - Lemmatizes tokens based on the specified language.
    """

    # Step 1: Convert text to lowercase
    text = text.lower()

    # Step 2: Clean the text
    text = clean_text(text, language)

    # Step 3: Tokenize the text
    tokens = word_tokenize(text)

    # Step 4: Remove stopwords and apply lemmatization
    if language == "english":
        # Remove English stopwords and lemmatize
        stop_words = set(stopwords.words("english"))
        tokens = [word for word in tokens if word not in stop_words]
        text = lemmatize_english(" ".join(tokens))

    elif language == "urdu":
        # Remove Urdu stopwords and lemmatize
        tokens = [word for word in tokens if word not in urdu_stopwords]
        text = lemmatize_urdu(" ".join(tokens))

    elif language == "punjabi":
        # Remove Punjabi stopwords and lemmatize
        tokens = [word for word in tokens if word not in punjabi_stopwords]
        text = lemmatize_punjabi(" ".join(tokens))

    # elif language == "siraiki":
    #     # Remove Siraiki stopwords and lemmatize
    #     tokens = [word for word in tokens if word not in siraiki_stopwords]
    #     text = lemmatize_siraiki(" ".join(tokens))

    # Return preprocessed text
    return text
```

###Apply Preprocessing and Prepare Dataset

```python
# Create a new dataframe where each review appears in all languages
multilingual_reviews = []

# Process each review in the original dataset
for _, row in df.iterrows():
    sentiment = row["Sentiment"]  # Sentiment of the review

    # Append preprocessed reviews in all languages
    multilingual_reviews.append({
        "text": preprocess_text(row["Summary"], "english"),
        "sentiment": sentiment,
        "language": "english",
    })
    multilingual_reviews.append({
        "text": preprocess_text(row["review_urdu"], "urdu"),
        "sentiment": sentiment,
        "language": "urdu",
    })
    multilingual_reviews.append({
        "text": preprocess_text(row["review_punjabi"], "punjabi"),
        "sentiment": sentiment,
        "language": "punjabi",
    })
    # multilingual_reviews.append({
    #     "text": preprocess_text(row["review_siraiki"], "siraiki"),
    #     "sentiment": sentiment,
    #     "language": "siraiki",
    # })

# print(multilingual_reviews)

# Convert the list to a DataFrame
multilingual_df = pd.DataFrame(multilingual_reviews)

# Convert to a Hugging Face Dataset
train_data = Dataset.from_pandas(multilingual_df)

# Display the dataset structure
print(train_data[0])
```

###Model Training

####XLM-RoBERTa

#####Tokenization with Hugging Face Tokenizer



```python
from transformers import XLMRobertaTokenizer

# Initialize the tokenizer for the multilingual model (XLM-RoBERTa)
tokenizer = XLMRobertaTokenizer.from_pretrained("xlm-roberta-base")

# Define tokenization function
def preprocess_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

# Tokenize the dataset
train_data = train_data.map(preprocess_function, batched=True)
```

```python
from transformers import XLMRobertaForSequenceClassification, Trainer, TrainingArguments

# Initialize the model with the appropriate number of sentiment labels
model = XLMRobertaForSequenceClassification.from_pretrained("xlm-roberta-base", num_labels=3)

training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    num_train_epochs=3,
    per_device_train_batch_size=4,
    save_steps=10_000,
    save_total_limit=2,
    report_to="none",  # Disable WandB logging
)

train_data = train_data.train_test_split(test_size=0.2)


# Create a Trainer instance
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_data["train"],
    eval_dataset=train_data["test"],  # Pass eval_dataset here
)

# Train the model
trainer.train()
```

```python
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments

# Initialize the tokenizer for mBERT
tokenizer = BertTokenizer.from_pretrained("bert-base-multilingual-cased")

# Define tokenization function
def preprocess_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

# Tokenize the dataset
train_data = train_data.map(preprocess_function, batched=True)

# Initialize the mBERT model with the appropriate number of sentiment labels
model = BertForSequenceClassification.from_pretrained("bert-base-multilingual-cased", num_labels=3)

# Define training arguments
# training_args = TrainingArguments(
#     output_dir="./results",
#     evaluation_strategy="epoch",  # Evaluate after each epoch
#     num_train_epochs=5,  # Start with 5 epochs
#     per_device_train_batch_size=8,  # Smaller batch size
#     save_steps=10_000,
#     save_total_limit=2,
#     report_to="none",  # Disable WandB logging
# )

training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="steps",  # Evaluate every X steps
    eval_steps=5000,              # Evaluate every 5000 steps
    num_train_epochs=3,           # Train for 3 epochs
    per_device_train_batch_size=16,  # Or 8 if memory is limited
    per_device_eval_batch_size=32,  # Larger batch size for evaluation
    learning_rate=3e-5,            # Suitable learning rate for fine-tuning
    warmup_steps=1300,             # 10% of total training steps as warm-up
    weight_decay=0.01,             # Regularization
    logging_dir="./logs",          # Directory for logs
    logging_steps=500,             # Log every 500 steps
    save_steps=5000,               # Save model every 5000 steps
    save_total_limit=2,            # Limit checkpoints to save space
    gradient_accumulation_steps=1, # Use 2 if per-device batch size is 8
    fp16=True,                     # Enable mixed precision training for faster computation
    seed=42,                       # For reproducibility
    report_to="none",  # Disable WandB logging
)

# Assume train_data is a DatasetDict
dataset_split = train_data.train_test_split(test_size=0.2)

# Access the train and eval splits
train_dataset = dataset_split["train"]
eval_dataset = dataset_split["test"]


# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,  # Use the training split
    eval_dataset=eval_dataset,    # Use the evaluation split
)

# Train the model
trainer.train()
```

###Evaluation and Fine-Tuning

```python
from sklearn.metrics import accuracy_score

# Get predictions
predictions = trainer.predict(eval_dataset)

# Extract logits and labels
logits = predictions.predictions
labels = predictions.label_ids

# Get predicted classes
predicted_classes = logits.argmax(axis=-1)

# Calculate accuracy
accuracy = accuracy_score(labels, predicted_classes)

print(f"Accuracy: {accuracy}")
```

```python
from transformers import XLMRobertaTokenizer, XLMRobertaForSequenceClassification
import torch

# Load the pre-trained tokenizer and model
tokenizer = XLMRobertaTokenizer.from_pretrained("xlm-roberta-base")
model = XLMRobertaForSequenceClassification.from_pretrained("xlm-roberta-base")

# Example long text
test_text_english = "Good flavor" # Example of a long text

# Function to split the input text into chunks of 512 tokens or less
def split_text(text, max_length=512):
    # Tokenize the text with padding and truncation
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding="max_length", max_length=max_length)
    return inputs

# Split the text into chunks (tokenized)
inputs = split_text(test_text_english)

# Ensure the model is in evaluation mode
device = torch.device("cuda")  # Change to "cuda" if using GPU
model = model.to(device)

# Convert inputs to tensor and move to the correct device
input_ids = inputs["input_ids"].to(device)
attention_mask = inputs["attention_mask"].to(device)

# Perform inference
with torch.no_grad():
    outputs = model(input_ids=input_ids, attention_mask=attention_mask)

logits = outputs.logits
predicted_class_idx = torch.argmax(logits, dim=-1).item()

# Mapping the predicted class to sentiment labels
label_mapping = {0: "negative", 1: "neutral", 2: "positive"}
predicted_sentiment = label_mapping.get(predicted_class_idx, "unknown")

print(f"The sentiment of the text is: {predicted_sentiment}")
```
